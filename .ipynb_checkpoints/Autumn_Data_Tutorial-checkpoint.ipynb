### Add a King county boudary

```python
#king_county_boundary_frame = gpd.read_file('C:/Users/aaron/Fun Mapping Work/Data/Maptime_Tutorial_Data/Washington_Libraries.geojson')
king_county_boundary_frame = gpd.read_file('C:/Users/aaron/Fun Mapping Work/Data/Maptime_Tutorial_Data/King_County_Political_Boundary_(no_waterbodies)___kingco_area.geojson')
```

```python
#king_county_boundary_frame.explore()
```

```python
### Make a king county frame

#geopandas.sjoin(left_df, right_df, how='inner', predicate='intersects', lsuffix='left', rsuffix='right', distance=None, on_attribute=None, **kwargs)

amanita_in_king_county_GDF = gpd.sjoin(amanita_geodataframe, king_county_boundary_frame, how = 'inner', predicate = 'intersects')
```

```python
amanita_in_king_county_GDF = amanita_in_king_county_GDF.drop('EDITDATE', axis = 1)
```

```python
#amanita_in_king_county_GDF.explore()
```

```python
amanita_in_king_county_GDF.head()
```

```python
hex_frame_kingco = amanita_in_king_county_GDF.h3.geo_to_h3_aggregate(7, operation = 'sum')
```

```python
hex_frame_kingco.explore(column = 'quantity', tooltip = tt)
```

```python
fig, ax = plt.subplots(figsize  = (21, 21))

hex_frame_kingco.plot(
                column = 'quantity',
                legend = True,
                norm=matplotlib.colors.LogNorm(vmin = 0.01, vmax = 100),
                ax = ax)

ax.set_title("King County Amanita Reports", fontsize=15)

```
